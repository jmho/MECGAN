```python
# example of loading the generator model and generating images
from math import sqrt
from numpy import asarray
from numpy.random import randn
import tensorflow as tf
import numpy as np
from tensorflow import keras
from keras.models import load_model
from matplotlib import pyplot

num_classes = 8
latent_dim = 128

# generate points in latent space as input for the generator
def generate_latent_points(latent_dim, n_samples, n_class):
    
    labels = keras.utils.to_categorical([n_class], num_classes)
    labels = tf.cast(labels, tf.float32)
    labels = tf.repeat(labels, repeats=n_samples)
    labels = tf.reshape(labels, (n_samples, num_classes))
    
    noise = tf.random.normal(shape=(1, latent_dim))
    noise = tf.repeat(noise, repeats=n_samples)
    noise = tf.reshape(noise, (n_samples, latent_dim))
    
    return tf.concat([noise, labels], 1)
 
# create and save a plot of generated images
def save_plot(examples, n_examples):
    # plot images
    examples = tf.cast(examples, tf.uint8)
    for i in range(n_examples):
        # define subplot
        pyplot.subplot(sqrt(n_examples), sqrt(n_examples), 1 + i)
        # turn off axis
        pyplot.axis('off')
        # plot raw pixel data
        pyplot.imshow(examples[i, :, :, :])
    pyplot.show()

def single_plot(data, path):
    # define subplot
    # turn off axis
    pyplot.axis('off')
    # plot raw pixel data
    data = tf.cast(data, tf.uint8)
    pyplot.imshow(data[0, :, :, :])
    pyplot.savefig(IMAGE_PATH)
    pyplot.close()
 
# load model
trained_gen = keras.models.load_model('generator_proto.h5')
latent_dim = 128
n_examples = 1 # must be a square
n_class =6 # dict_keys(['amusement', 'anger', 'awe', 'contentment', 'disgust', 'excitement', 'fear', 'sadness'])

label_dict = ['amusement', 'anger', 'awe', 'contentment', 'disgust', 'excitement', 'fear', 'sadness']

# generate images
noise_and_labels = generate_latent_points(latent_dim, n_examples, n_class)
# generate images
@tf.function
def pred(noise_and_labels):
    return trained_gen(noise_and_labels)
# scale from [-1,1] to [0,1]
X = pred(noise_and_labels)
X *= 255.0
print(X.shape)
# plot the result

IMAGE_PATH = "original.png"

print(label_dict[n_class])
save_plot(X, n_examples)
single_plot(X, IMAGE_PATH)
```

```python
import PIL
import tensorflow_hub as hub
import matplotlib.pyplot as plt
import time
```

```python
SAVED_MODEL_PATH = "https://tfhub.dev/captain-pool/esrgan-tf2/1"


def plot_image(image, title=""):
  """
    Plots images from image tensors.
    Args:
      image: 3D image tensor. [height, width, channels].
      title: Title to display in the plot.
  """
  image = np.asarray(image)
  image = tf.clip_by_value(image, 0, 255)
  image = PIL.Image.fromarray(tf.cast(image, tf.uint8).numpy())
  plt.imshow(image)
  plt.axis("off")
  plt.title(title)
```

```python
plot_image(tf.squeeze(X), title="Original Image")
```

```python
model = hub.load(SAVED_MODEL_PATH)
```

```python
start = time.time()
fake_image = model(X)
print("Time Taken: %f" % (time.time() - start))

plot_image(tf.squeeze(fake_image), title="Super Res Image")
```

```python
print(fake_image.shape)
fake_image = tf.squeeze(fake_image)
```

```python
def save_image(image, filename):
  """
    Saves unscaled Tensor Images.
    Args:
      image: 3D image tensor. [height, width, channels]
      filename: Name of the file to save.
  """
  if not isinstance(image, PIL.Image.Image):
    image = tf.clip_by_value(image, 0, 255)
    image = PIL.Image.fromarray(tf.cast(image, tf.uint8).numpy())
  image.save("%s.jpg" % filename)
  print("Saved as %s.jpg" % filename)
```

```python
save_image(fake_image, "Fake-image")
```
